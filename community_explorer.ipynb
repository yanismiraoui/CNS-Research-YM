```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
roi_coords = pd.read_csv('./xGW-GAT/community_networks/roi_coords.csv', header=None)
# inverse the row and column
roi_coords = roi_coords.T
# set the column names
roi_coords.columns = ['name', 'x', 'y', 'z']
# drop the first row
roi_coords = roi_coords.drop([0])
# set index to name
roi_coords = roi_coords.set_index('name')
# convert 3 cols to float
roi_coords = roi_coords.astype(float)
```

```python
roi_coords
```

```python
# plot the coordinates of the ROIs in a 3D space
%matplotlib widget
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(roi_coords['x'].values, roi_coords['y'].values, roi_coords['z'].values)
plt.show()
```
