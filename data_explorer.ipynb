## Data from ABIDE

```python
import os
import scipy.io as sio
import numpy as np
import torch
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nilearn import plotting
```

```python
# Loading files
folders = os.listdir('./data_abide/ABIDE_pcp/cpac/filt_noglobal')
store = []
errors = 0
subject_ids = []

print("Number of folders: ", len(folders))
for f in folders:
    files = os.listdir('./data_abide/ABIDE_pcp/cpac/filt_noglobal/' + f)
    for fi in files:
        if fi.endswith('.mat') and '_correlation' in fi and 'partial' not in fi:
            try:
                mat = sio.loadmat('./data_abide/ABIDE_pcp/cpac/filt_noglobal/' + f + '/' + files[0])
                store.append(mat['connectivity'])
                subject_ids.append(f)
            except Exception as e:
                errors += 1
                continue

print("Number of correctly loaded files: ", len(store))
print("Number of errors loading the file: ", errors)
print("Size of correlation matrix: ", store[0].shape)
print("Total number of subjects: ", len(store))
```

```python
# Plot the correlation matrix for the last subject
plotting.plot_matrix(mat['connectivity'], colorbar=False,
                         title='Correlation last subject')
```

```python
# Transform list of correlation matrices into a 3D array
correlation_matrices = np.array(store)
print("Shape of correlation matrices (3D array): ", correlation_matrices.shape)
```

```python
DATA_FOLDER = './data_abide/xGW-GAT/'
# conn = np.transpose(correlation_matrices, (2, 1, 0))
conn = torch.Tensor(correlation_matrices)
print(f"Connectome (X) shape: {conn.shape}") # n_ROIs, n_ROIs, n_subjects

torch.save(conn, f"{DATA_FOLDER}/connectomes_0.npy")
```

```python
# Load csv with labels
df = pd.read_csv(f"{DATA_FOLDER}/ABIDE_pcp/Phenotypic_V1_0b_preprocessed1.csv", index_col=0)
df
```

```python
# Label is DX_GROUP
# keep only the rows that have a 'subject' in the folders
# transform to string
df["subject_str"] = df["subject"].apply(lambda x: str(x))
sub_df = df[df["subject_str"].isin(subject_ids)]
sub_df
```

```python
sub_df["DX_GROUP"].value_counts()
```

```python
scores = torch.LongTensor(sub_df["DX_GROUP"].values)
print(f"Gait scores (y) shape: {scores.shape}") # n_subjects
print(scores)
torch.save(scores, f"{DATA_FOLDER}/scores.npy")
```

```python
merged_df_multi = pd.DataFrame(list(zip(correlation_matrices, scores, sub_df["subject"])),
               columns =['corr', 'labels', 'p_ids'])
records_multi = merged_df_multi.to_records(index=False)
data_multi = np.array(records_multi, dtype=records_multi.dtype.descr)

np.save(f"{DATA_FOLDER}/private.npy", data_multi)
```

## Loading as example to get 1009 subjects

```python
# load abide.npy
data = np.load(f"./data_abide/ABIDE_pcp/abide.npy",allow_pickle=True).item()
# keeping only 100 subjects randomly
random_ids = np.random.choice(len(data['label']), 100, replace=False)
data = {k: v[random_ids] for k, v in data.items()}
print("Number of subjects: ", len(data['label']))
data.keys()
```

```python
DATA_FOLDER = './data_abide/xGW-GAT/'
# conn = np.transpose(correlation_matrices, (2, 1, 0))
conn = torch.Tensor(data["corr"])
print(f"Connectome (X) shape: {conn.shape}") # n_ROIs, n_ROIs, n_subjects

torch.save(conn, f"{DATA_FOLDER}/connectomes_0_small.npy")
```

```python
scores = torch.LongTensor(data["label"])
print(f"Gait scores (y) shape: {scores.shape}") # n_subjects
print(scores)
torch.save(scores, f"{DATA_FOLDER}/scores_small.npy")
```

```python
merged_df_multi = pd.DataFrame(list(zip(data['corr'], scores, [i for i in range(len(scores))])),
               columns =['corr', 'labels', 'p_ids'])
records_multi = merged_df_multi.to_records(index=False)
data_multi = np.array(records_multi, dtype=records_multi.dtype.descr)

np.save(f"{DATA_FOLDER}/private_small.npy", data_multi)
```
