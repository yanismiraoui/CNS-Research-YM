```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data_abide_aal/ABIDE_pcp/Phenotypic_V1_0b_preprocessed1.csv', low_memory=False, index_col=0)
```

```python
df
```

```python
# Plot DX_GROUP as a histogram using seaborn with nice blue (#007c92) and violet (#610059) colors
sns.set_style('white')
# Transparent background
sns.set_style('whitegrid', {'axes.facecolor': '.9'})
plt.figure(figsize=(12, 6))
# Make text bigger by choosing font scale 1.5
plt.rc('font', size=18)
sns.countplot(x='DX_GROUP', data=df, palette=['#007c92', '#610059'])
plt.xlabel('Diagnosis')
# Change x axis labels to something more readable (ASD and Control)
plt.xticks([0, 1], ['Control', 'ASD'])
plt.ylabel('Number of Subjects')
plt.title('Number of Subjects with ASD and Controls in ABIDE')
plt.show()
```
